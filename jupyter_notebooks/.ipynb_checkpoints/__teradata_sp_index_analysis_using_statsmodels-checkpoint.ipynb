```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from dateutil.relativedelta import relativedelta
import seaborn as sns
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

# Load S&P Index data from local HBase

```python
import happybase

# Since I already put the S&P data into local HBase, now I can directly read it
connection = happybase.Connection('localhost')
hbase_table_name = "Table_YAHOO_INDEX_GSPC"
hbase_table = connection.table(hbase_table_name)
```

```python
series = []
for key, data in hbase_table.scan():
    series.append(data)
    
sorted_items_by_date = sorted(series, key=lambda d: d['cf:DATE'])
# print the first 50 indices with dates (they are ordered)
print sorted_items_by_date[:50]
```

```python
# print the first 50 indices
sorted_series = [d['cf:ADJUSTED_PRICE'] for d in sorted_items_by_date]
print sorted_series[:50]
```

# Convert two interested attributes into pandas dataframe

```python
dates = []
adjusted_prices = []
for d in sorted_items_by_date:
    dates.append(d['cf:DATE'])
    adjusted_prices.append(float(d['cf:ADJUSTED_PRICE']))
    
df = pd.DataFrame({'date': dates, 'adjusted_price': adjusted_prices})
```

# Only using partial data starting from 1997

```python
df = df[12000:]
df
```

# Visualize the S&P indices since 1997

```python
df['index'] = pd.to_datetime(df.date)
df.set_index(['index'], inplace=True)
df.index.name=None
df.adjusted_price.plot(figsize=(10, 6), title= 'S&P Index', fontsize=14)
```

# Seasonal analysis and visualization

```python
# it is not easy for me to know what is the season number, so I just guessed 5 days
decomposition = seasonal_decompose(df.adjusted_price, freq=5)
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
```

```python
# what about 20 days?
decomposition = seasonal_decompose(df.adjusted_price, freq=20)
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
```

# Check stationarity of time series

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(df.adjusted_price)
```

```python
df.adjusted_price_log= df.adjusted_price.apply(lambda x: np.log(x))  
test_stationarity(df.adjusted_price_log)
```

# Perform difference operations due to in-stationarity

```python
df['first_difference'] = df.adjusted_price - df.adjusted_price.shift(1)  
test_stationarity(df.first_difference.dropna(inplace=False))
```

```python
df['log_first_difference'] = df.adjusted_price_log - df.adjusted_price_log.shift(1)  
test_stationarity(df.log_first_difference.dropna(inplace=False))
```

```python
df['seasonal_difference'] = df.adjusted_price - df.adjusted_price.shift(5)  
test_stationarity(df.seasonal_difference.dropna(inplace=False))
```

```python
df['log_seasonal_difference'] = df.adjusted_price_log - df.adjusted_price_log.shift(5)  
test_stationarity(df.log_seasonal_difference.dropna(inplace=False))
```

# Get ACF and PACF

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df.log_seasonal_difference.iloc[6:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df.log_seasonal_difference.iloc[6:], lags=40, ax=ax2)
```

# Fit seasonal ARIMA model

```python
mod = sm.tsa.statespace.SARIMAX(df.adjusted_price, trend='n', order=(0,1,0), seasonal_order=(0,1,1,5))
results = mod.fit()
print results.summary()
```

```python
mod = sm.tsa.statespace.SARIMAX(df.adjusted_price, trend='n', order=(0,1,0), seasonal_order=(1,1,1,5))
results = mod.fit()
print results.summary()
```

# Forecast index

```python
#df['forecast'] = results.predict(start = 102, end= 114, dynamic= True)  
df['forecast'] = results.predict(start = 3000, end= 3200, dynamic= True)  
df[['adjusted_price', 'forecast']].plot(figsize=(20, 10)) 
plt.savefig('ts_df_predict.png', bbox_inches='tight')
```

The green color shows the predicted indices in 200 days, just for the purpose of visualization. Otherwise, it will be hard for visual inspection. Note that this is simply a demonstration of the feasibility of forcasting
future indices using the statsmodel library. I am not indended to produce any accurate models for prediction here.
